## Create d1

```
d1 = pd.DataFrame({'city':['Chandigarh','Delhi','Kanpur','Chennai','Manali'],'Temperature':[15,22,20,26,-2]})
print(d1)
```

## Shape of d1

```
print(d1.shape)
```

## City column

```
city = d1['city']
print(city)
print(type(city))
```

## Create d2

```
d2 = pd.DataFrame({'city':['Bengaluru','Coimbatore','Srirangam','Pondicherry'],'Temperature':[24,35,36,39]})
print(d2.shape)
```

## Merge d1 and d2

```
d3 = pd.concat([d1,d2])
print(d3)
```

## Temp <=20

```
print(d3[d3['Temperature']<=20])
print(len(d3[d3['Temperature']<=20]))
```

## Temp >=35

```
print(d3[d3['Temperature']>=35])
```

## Union Territory column

```
def is_ut(x):
    return x in ['Chandigarh','Delhi','Pondicherry']

d3['is_ut'] = d3['city'].apply(is_ut)
print(d3)
```

## Celsius to Fahrenheit

```
def c_to_f(x):
    return (x*9/5)+32

print(c_to_f(10))
d3['temp_f'] = d3['Temperature'].apply(c_to_f)
print(d3)
```

## Select UT cities

```
print(d3[d3['is_ut']==True])
```

## Temp > 90F

```
print(d3[d3['temp_f']>90])
```

## First 3 rows

```
print(d1.head(3))
```

## Last two columns

```
print(d1.iloc[:, -2:])
```

## Groupby

```
data = {'Category':['A','A','B','B','C'],'Values':[10,20,30,40,50]}
df2 = pd.DataFrame(data)

print(df2.groupby('Category').sum())
print(df2.groupby('Category').count())
print(df2.groupby('Category').min())
print(df2.groupby('Category').max())
```
