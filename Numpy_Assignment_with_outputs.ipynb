## 3. Dimensions, Shape, Data Type

```python
a=np.array(10)
b=np.array([1,2,3])
c=np.array([[1,2,3],[4,5,6]])
d=np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])

print(a.ndim,b.ndim,c.ndim,d.ndim)
print(a.shape,b.shape,c.shape,d.shape)
print(c.dtype,d.dtype)
print(type(a),type(b))
print(len(b))
```
